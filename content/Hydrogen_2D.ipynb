```python
# This code solves the Schrdinger equation for a hydrogen atom in 2D
# with a Coulomb interaction of V(r) = -2/r.
# It calculates the binding energies and wave functions
# for different partial wave quantum numbers m_angular.

# Author: Mohammadreza Hadizadeh (mhadizadeh@centralstate.edu)
# February 2025.

# Please cite the companion article if you use this resource:
# K. Adderley, K. Mohseni, and M. R. Hadizadeh,
# Open-source interactive website for exciton modeling via the LippmannSchwinger equation (2025),
# URL: https://csu-physics.github.io/2B_2D/.

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from scipy.special import roots_legendre
from scipy.sparse.linalg import eigs

# ----------------------------------------------------------------------
# Global numerical parameters and physical constants
# ----------------------------------------------------------------------
Np = 200             # Number of momentum grid points (radial momentum p)
Nphi = 40            # Number of angular points (phi) for angle integration
pi = np.pi           # Convenience alias for pi
ONE = 1.0            # Convenience constant for 1.0 (used for clarity)
ZERO = 0.0           # Convenience constant for 0.0 (unused here but common in numerics)
Tol_lambda = 1.0e-10 # Convergence tolerance for eigenvalue Lambda  1  
mass = 1.0           # Particle mass in chosen units
m_angular = 0        # Orbital quantum number (azimuthal angular momentum m)

# ----------------------------------------------------------------------
# Matplotlib style tweaks for more publication-ready plots
# ----------------------------------------------------------------------
plt.rcParams.update({
    "figure.figsize": (6, 4),
    "axes.labelsize": 14,
    "axes.titlesize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "legend.fontsize": 12,
    "lines.linewidth": 2,
    "axes.grid": False,
    "figure.dpi": 150,
})

# -----------------------
# Subroutines and Functions
# -----------------------
def hyperbolic(Np):
    """
    Generate hyperbolic distribution points and their weights using Legendre roots.
    Returns:
        (P, dP): Tuple of numpy arrays of shape (Np,)
                 P is the transformed momentum points
                 dP is the corresponding weights
    """
    # X: standard GaussLegendre points in [-1, 1]
    # dX: corresponding GaussLegendre weights
    X, dX = roots_legendre(Np)

    # Map Legendre points X  (-1, 1) to a semi-infinite momentum domain P  (0, )
    # via the transformation P = (1 + X) / (1 - X).
    P = (1.0 + X) / (1.0 - X)

    # Compute the Jacobian of the transformation to get new weights dP
    dP = 2.0 / (1.0 - X)**2 * dX
    return P, dP

def linear(Nphi, a, b):
    """
    Generate linear distribution points and their weights for variables over an interval [a, b].
    Args:
        Nphi (int): Number of points in phi.
        a (float): Start of the interval.
        b (float): End of the interval.

    Returns:
        (x, dx): Tuple of numpy arrays of shape (Nphi,)
                 x is the linearly mapped points from [-1, 1] to [a, b]
                 dx is the corresponding weights
    """
    # Get GaussLegendre points and weights on [-1, 1]
    X, dX = roots_legendre(Nphi)

    # Affine map X  [-1, 1]  x  [a, b]
    x = (b - a) / 2 * X + (b + a) / 2

    # Scale weights by the interval length
    dx = (b - a) / 2 * dX
    return x, dx

def eigenvalue(Mmatrix):
    """
    This function calculates a few eigenvalues and eigenvectors of the matrix Mmatrix.

    Args:
        Mmatrix (numpy.ndarray): A square matrix from which to find eigenvalues and eigenvectors.

    Returns:
        WF (numpy.ndarray): The eigenvectors corresponding to the largest eigenvalue.
        Lambda (float): The largest eigenvalue.
    """
    # Determine the size (dimension) of the matrix
    dim_Mmatrix = Mmatrix.shape[0]

    # Specify the number of eigenvalues and eigenvectors to find
    # NEV: number of eigenvalues,
    # NCV: number of Arnoldi vectors, must be greater than NEV
    NEV = 10
    NCV = 2 * NEV # A larger NCV can accelerate convergence

    # WHICH specifies which eigenvalues to target:
    # 'LR': Largest Real part; useful when interested in eigenvalues near 1 on real axis
    WHICH = 'LR'

    # Use ARPACK (via scipy.sparse.linalg.eigs) to find NEV eigenvalues/eigenvectors of Mmatrix
    eigenvalues, eigenvectors = eigs(Mmatrix, k=NEV, which=WHICH, ncv=NCV)

    # Target eigenvalue near 1 (fixed point of the LippmannSchwinger equation)
    target_eigenvalue = 1.0
    # Find index of eigenvalue closest to 1 in magnitude
    idx_min = np.argmin(np.abs(eigenvalues - target_eigenvalue))

    # Extract that eigenvalue and corresponding eigenvector
    Lambda = eigenvalues[idx_min]
    WF = np.abs(eigenvectors[:, idx_min])  # Take absolute value to remove arbitrary overall phase

    return WF, Lambda

def compute_V_2B(Np, Nphi, m_angular, p, xp, phi, dphi):
    """
    Computes the 3D array V_2B (for potential calculations) and integrates it
    over phi to form a 2D array V_2B_2d.

    Args:
        Np (int): Number of p points
        Nphi (int): Number of phi points
        m_angular: state quantum number
        p (np.ndarray): 1D array of momentum values of length Np
        xp (np.ndarray): 1D array of cos(phi) values of length Nphi
        dphi (np.ndarray): 1D array of differential phi weights of length Nphi

    Returns:
        V_2B_2d (np.ndarray): 2D array of shape (Np, Np), the integrated potential over phi
    """
    # Initialize the V_2B 3D array:
    #   V_2B[ip, ipp, ixp]  V(p, p', ) for given p[ip], p[ipp], and angle [ixp]
    V_2B = np.zeros((Np, Np, Nphi))

    # Populate the V_2B array with the momentum-space Coulomb interaction:
    # q is the momentum transfer |p - p'| in 2D
    for ip in range(Np):
        for ipp in range(Np):
            for ixp in range(Nphi):
                # Law of cosines in momentum space for magnitude of q = |p - p'|
                q = np.sqrt(p[ip]**2 + p[ipp]**2
                            - 2.0 * p[ip] * p[ipp] * xp[ixp])
                # Potential in momentum space, here V(q) = -1 / (q * pi)
                V_2B[ip, ipp, ixp] = -1.0 / (q * pi)

    # Integrate V_2B over the angle , including angular dependence via cos(m )
    # to project onto the m_angular partial wave.
    V_2B_2d = np.zeros((Np, Np))
    # Precompute cos(m * ) factor for all angular grid points
    cos_terms = np.cos(m_angular * phi)  # This computes the cosine term for all phi values

    # Loop over momentum indices to perform angular integration
    for ipp in range(Np):
        for ip in range(Np):
           # phi integration: sum over  with GaussLegendre weights dphi
           V_2B_2d[ip, ipp] = np.sum(dphi * V_2B[ip, ipp, :] * cos_terms)

    return V_2B_2d

def LS_2D(Np, pp, dp, V_NR, BE2_1, BE2_2):
    """
    Demonstration routine for computing eigenvalues using the 2D LS approach.

    Args:
        Np (int): Number of momentum points
        pp (np.ndarray): Momentum array of shape (Np,)
        dp (np.ndarray): Weights array associated with pp
        V_NR (np.ndarray): 2D potential array V_NR of shape (Np, Np)

    Returns:
        (BE2, WF): Tuple of (float, np.ndarray)
                   BE2 is the last computed energy
                   WF is the corresponding wave function array of shape (Np,)
    """
    # Number of iterations/energy guesses in the LippmannSchwinger search
    Ne = 20

    # Arrays to store binding energy guesses and corresponding eigenvalues Lambda
    BE_ary = np.zeros(Ne)
    Lam_ary = np.zeros(Ne)

    # Initial two energy guesses supplied by the caller
    BE_ary[0] = BE2_1
    BE_ary[1] = BE2_2

    # Iteratively refine the binding energy BE2 using an extrapolation scheme
    for ie in range(Ne):
        # Extrapolation-like step if more than two previous points are available:
        # This uses a linear extrapolation based on how Lambda moves toward 1.
        if ie > 1:
            BE_ary[ie] = (BE_ary[ie-1]
                          + (1 - Lam_ary[ie-1])/(Lam_ary[ie-2] - Lam_ary[ie-1])
                          * (BE_ary[ie-2] - BE_ary[ie-1]))
        # Current binding energy guess
        BE2 = BE_ary[ie]

        # Build the kernel matrix Mmatrix for the LS equation in momentum space:
        #   M(ip, ipp) ~ dp * p' * V(p, p') / (E - p^2/m)
        Mmatrix = np.zeros((Np, Np))
        for ip in range(Np):
            for ipp in range(Np):
                # Denominator: free Green's function in momentum space
                denom = (BE2 - pp[ip]**2 / mass)
                if denom != 0:
                    Mmatrix[ip, ipp] = (dp[ipp] * pp[ipp] * V_NR[ip, ipp] / denom)
                # If denom were exactly zero, we would skip (avoids division by zero).

        # Solve the eigenvalue problem M WF = Lambda WF
        WF, Lambda = eigenvalue(Mmatrix)

        # Normalize WF with respect to the radial momentum measure dp * p
        WF /= np.sqrt(np.sum(WF**2 * dp * pp))

        # Store the real part of Lambda for extrapolation/convergence check
        Lam_ary[ie] = np.real(Lambda)

        # Print diagnostic information for each search iteration
        print(f"Search {ie+1} | BE2 = {BE2}, Lambda = {Lambda}")

        # Convergence check: if Lambda is sufficiently close to 1, stop iterating
        if np.abs(Lambda - 1.0) < Tol_lambda:
            break

    # Return the last binding energy guess and corresponding (normalized) wave function
    return BE2, WF

# -----------------------
# Main Program
# -----------------------
def main():
    # Set up momentum grid p and corresponding integration weights dp
    p, dp = hyperbolic(Np)

    # Set up angular grid phi  [0, 2] and its integration weights dphi
    phi, dphi = linear(Nphi, 0, 2.0 * pi)

    # xp = cos(phi) will be used in computing the momentum transfer q
    xp = np.cos(phi)

    # Store wave functions for different principal quantum numbers n
    WF_matrix = []

    # Loop over values of n from m_angular up to 4 (since range upper bound is 5)
    # In 2D Coulomb problem, n labels bound states (analog of principal quantum number).
    for n in range(m_angular, 5):
        # Exact analytical energy for 2D Coulomb-like problem (in chosen units)
        E_exact = -ONE / (n + 0.5)**2  # Calculate the exact energy for each n

        # Initial lower and upper guesses for the binding energy (bracketing)
        BE2_1 = E_exact * 0.95  # Lower bound for energy estimate
        BE2_2 = E_exact * 1.05  # Upper bound for energy estimate

        # Compute the partial-wave projected potential matrix V_2B_2d for given m_angular
        V_2B_2d = compute_V_2B(Np, Nphi, m_angular, p, xp, phi, dphi)

        # Solve LS equation to obtain numerical binding energy BE2 and wave function WF(p)
        BE2, WF = LS_2D(Np, p, dp, V_2B_2d, BE2_1, BE2_2)

        # Relative percentage difference between numerical and exact energies
        rel_error = np.abs((BE2 - E_exact) / E_exact) * 100.0  # in %

        # Store wave function for later comparison plots
        WF_matrix.append(WF)

        # Print energies and relative percentage difference for this state
        print('')
        print(f'For n = {n}:')
        print(f'Computed Energy: {BE2}')
        print(f'Exact Energy:    {E_exact}')
        print(f'Relative Percentage Difference: {rel_error:.4f} %\n')

        # ------------------------------------------------------------------
        # Plot |(p)| vs. momentum for this (n, m_angular)  publication ready
        # ------------------------------------------------------------------
        fig, ax = plt.subplots()
        ax.loglog(p, np.abs(WF), label=f'm={m_angular}, n={n}')
        ax.set_xlabel(r'Relative momentum $p$')
        ax.set_ylabel(r'Wave function $|\Psi(p)|$')
        # ax.set_title(f'Wave Function vs. Momentum for n={n}, m={m_angular}')
        ax.legend(frameon=False)
        fig.tight_layout()
        plt.show()

    # ----------------------------------------------------------------------
    # Combined plot: comparison of |(p)| for different n at fixed m_angular
    # ----------------------------------------------------------------------
    fig, ax = plt.subplots()
    for i, WF in enumerate(WF_matrix):
        ax.loglog(p, np.abs(WF), label=f'n={i + m_angular}')
    ax.set_xlabel(r'Relative momentum $p$')
    ax.set_ylabel(r'Wave function $|\Psi(p)|$')
    ax.set_title(rf'Comparison of $|\Psi(p)|$ for Different $n$ (m={m_angular})')
    ax.legend(frameon=False)
    fig.tight_layout()
    plt.show()

# Standard Python entry point: only run main() when executed as a script
if __name__ == "__main__":
    main()
```
